## Recursion
Recursion is a process in which a function solves a problem by calling itself on smaller subproblems.

Instead of using a loop, the function keeps calling itself with a smaller version of the problem until a stopping condition is reached.

#### Two Mandatory parts of recursion: (Every recursion must have this)
1) Base case:
    Condition where recursion stops. Without a base, the recursion never ends.
2) Recursive Case:
    Function calls itself with a smaller problem.


<img src=".\Reference Images\recursion.png" width="1000">


```python
# Example of Recursion: Count Down
def countdown(n):
    if n==0:    # Base case. Without it, recursion never ends
        return
    print(n)
    countdown(n-1)   # Recursive call

countdown(5)
```
