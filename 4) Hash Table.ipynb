### 4) HASH TABLE:
A hash table is a data structure that uses a hash function to map keys to storage locations, allowing very fast key-based lookup, insertion, update, and deletion. In Python, the dict data type is a hash table implementation.

So A hash function is a function that takes an input (called a key) and converts it into a fixed-size integer value called a hash value or hash code which can be used to efficiently store and retrieve data in a hash table.

#### Program for Hash Table

```python
# WRITING OUR OWN HASH FUNCTION.
# Note: We are not handling collisions in this code. We will learn it later.

class HashTable:     #create a custom hashtable class
    def __init__(self):    # constructor
        self.Max = 20     # hash table size = 20 (indexes 0 to 19)
        self.arr = [None for i in range(self.Max)]  # creates [None, None, None, None, ...] 100 times or based on table size.

    def get_hash(self,key):  # hash function - convert a key into an array index
        h = 0
        for char in key:    # it will loop through all characters in the key and converts them into ASCII values.
            h += ord(char)  # ord() converts characters to ASCII values.
        return h % self.Max  # Modulo operation to generate final hash index

# e.g. ord(a) is 97. It will add all ASCII values generated from key.

    def add(self,key,val):   # store value in the array
        h = self.get_hash(key)   # Generate hash
        self.arr[h] = val        # store value at specific index as per hash

    def get(self,key):      # retrieve value
        h = self.get_hash(key)   # hash again
        return self.arr[h]       # retrieve the value using hash

```

```python
t = HashTable()
t.add('march 6',130)     #Store key value pair in array
print(t.get_hash('march 6'))  #print hash value of march 6
```

```python
# when we print the array, we will see value of 'march 6' key is stored at 9th index
print(t.arr)        # print array (length - 20)
print(t.get('march 6'))
```

 Same as previous code, but it uses Python special methods (magic methods).
Special Methods allow your custom object to behave like built-in Python objects. It makes dictionary style syntax possible.

```python
class HashTable:
    def __init__(self):
        self.Max = 20
        self.arr = [None for i in range(self.Max)]

    def get_hash(self,key):
        h = 0
        for char in key:
            h += ord(char)
        return h % self.Max

    def __setitem__(self,key,val):
        h = self.get_hash(key)
        self.arr[h] = val

    def __getitem__(self,key):
        h = self.get_hash(key)
        return self.arr[h]

    def __delitem__(self,key):
        h = self.get_hash(key)
        self.arr[h] = None

```

#### COLLISION
A collision happens when two different keys map to the same bucket/index in a hash table.

Why Collisions Happen? - Since a hash table has limited buckets.

Example: Buckets: 0 to 9

But there can be thousands of keys. Different keys may end up producing the same bucket index.

#### Hash Table Program Using Collision Handling (Chaining Method)

```python
class HashTable:
    def __init__(self):
        self.MAX = 10
        # Create a list with 10 empty buckets. Each bucket is itself a list that can store multiple key-value pairs.
        # This is called SEPARATE CHAINING and is used to handle collisions.
        self.arr = [[] for i in range(self.MAX)]

    def get_hash(self, key):
        # Convert a string key into an array index.
        hash = 0
        # Add ASCII values of all characters in the key.
        for char in key:
            hash += ord(char)
        # Compress the large number into a valid bucket index (0-9).
        return hash % self.MAX

    def __getitem__(self, key):
        # Runs automatically when: value = t["march 6"]
        # Find bucket index for the key.
        arr_index = self.get_hash(key)
        # Search all key-value pairs in that bucket.
        for kv in self.arr[arr_index]:
            # kv[0] = key
            # kv[1] = value
            if kv[0] == key:
                return kv[1]

    def __setitem__(self, key, val):
        # Runs automatically when: t["march 6"] = 130
        # Find bucket index.
        h = self.get_hash(key)
        # Flag to track whether key already exists.
        found = False

        # Loop through all elements in the bucket.
        for idx, element in enumerate(self.arr[h]):
            # element looks like: ("march 6", 130)
            # Check if current key already exists.
            if len(element) == 2 and element[0] == key:
                # Update existing value.
                self.arr[h][idx] = (key, val)
                found = True
        # If key was not found, insert a new key-value pair.
        if not found:
            self.arr[h].append((key, val))

    def __delitem__(self, key):
        # Runs automatically when: del t["march 6"]
        # Find bucket index.
        arr_index = self.get_hash(key)
        # Search bucket for matching key.
        for index, kv in enumerate(self.arr[arr_index]):
            # kv[0] = key
            # kv[1] = value
            if kv[0] == key:
                print("del", index)
                # Remove the key-value pair from the bucket.
                del self.arr[arr_index][index]
```

```python
t = HashTable()
print(t.arr)       # Print empty array
t["march 6"] = 310
t["march 7"] = 420
t["march 8"] = 67
t["march 17"] = 63457

print(t.arr)
```

### Exercise- 1

nyc_weather.csv contains new york city weather for first few days in the month of January. Write a program that can answer following,
- What was the average temperature in first week of Jan
- What was the maximum temperature in first 10 days of Jan

```python
#
nyc_weather = {}  # empty dict
with open("D:/GithubRepos/DSA_in_Python/Datasets/nyc_weather.csv","r") as f:        # read the file from given path. Using with, it closes the file after loop
    next (f)            # to skip the first line which contains header.
    for line in f:
        tokens = line.split(',')           #csv is a delimited text file where values are separated by a delimiter, which is usually a comma(,)
        date = tokens[0]
        temp = int(tokens[1])
        nyc_weather[date] =temp     # dict do not have append method, so each execution add a new pair in the dict.

print(nyc_weather)

```

```python
# Average temperature based on number of days
def average_temp(day_count):
    if day_count <= 0:                       # One more thing we can add if day_count is greater than the length of dictionary
        raise Exception("Day count must be greater than 0")
    if day_count > len(nyc_weather):
        raise Exception("Day count is greater than the available data duration")
    avg_temp = 0
    count = 0
    for value in nyc_weather.values():
        if count == (day_count):
            break
        avg_temp += value
        count += 1

    return avg_temp/day_count

print(average_temp(7))
```

```python
# Maximum temperature in first n days
def max_temp(day_count):
    if day_count <= 0:
        raise Exception("Day count must be greater than 0")
    if day_count > len(nyc_weather):
        raise Exception("Day count is greater than the available data duration")
    max_temp = next(iter(nyc_weather.values()))   #assigning 1st temperature to max value initially
    count = 0
    for value in nyc_weather.values():
        if count == (day_count):
            break
        if value > max_temp:
            max_temp = value
        count += 1
    return max_temp

print(max_temp(10))   #max temp in first 10 days
```
