## Array
An array is a collection of elements stored in contiguous(adjacent) memory locations. Almost every advanced data structure starts from arrays.

In Python, there is no dedicated built-in "array" data type used in the same way as C or Java arrays. However, several Python data structures behave like arrays and are commonly used in DSA.

we mostly use: "list" as dynamic arrays. Each element has value and index(position).

We have a similar data structure in python called tuple similar to list, but it is not considered an array.
- Python List: An array of pointers that over-allocates space so it can dynamically grow, shrink, and re-arrange elements on the fly.
-  Tuple: A fixed-size block of pointers that can never change its size, shape, or contents after birth. It lacks the scaling logic that makes an array dynamic.

#### Why Arrays are Important

Arrays are one of the most fundamental data structures because:
1) Fast access to elements.
2) Efficient memory usage.
3) Basis for many advanced data structures such as: Lists, Stacks,Queues,Heaps,Hash Tables,Matrices

Traditionally arrays store same-type elements. But in Python lists can contain mixed types, but in DSA theory arrays are usually homogeneous.

##### Basic Operations on Array

```python
# Traversal - visiting every element in array -> Time Complexity:O(n)
arr = [10,20,30]
for x in arr:
    print(x)
```

```python
# Access from array -> Time Complexity: O(1)
print(arr[2])
```

```python
# Search
# Linear Search - Time Complexity: O(n)
for i in arr:
     if i == 30:
         print("found")
```

```python
# Insertion:
arr.append(40)  # append at the end of the list - Time complex O(1)
arr.insert(2,50) # insert 50 at 2nd index. - Time complex O(n)
# Note: Worst case is insert at beginning
print(arr)
```

```python
# Deletion
del arr[2]  # Delete value at index 2 - Time complex O(n)
# Time complexity O(n), since remaining elements will shift.

arr.remove(30) # Removing an item by its value
print(arr)
```

#### Exercise-1
Create a list to store monthly expenses and using that find out,

January - 2200, February - 2350, March - 2600, April - 2130, May - 2190

1. In Feb, how many dollars you spent extra compare to January?
2. Find out your total expense in first quarter (first three months) of the year.
3. Find out if you spent exactly 2000 dollars in any month
4. June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list
5. You returned an item that you bought in a month of April and
got a refund of 200$. Make a correction to your monthly expense list
based on this

```python
# In Feb, how many dollars you spent extra compare to January?
expense = [2200,2350,2600,2130,2190]
print("Expense list month-wise (Jan-May):", expense)
print("Extra money spent in Feb as compare to Jan is: ",expense[1]-expense[0],"$")
```

```python
# Find out your total expense in first quarter (first three months) of the year.
print("Q1 expense:",expense[0]+expense[1]+expense[2],"$")
```

```python
# Find out if you spent exactly 2000 dollars in any month
# METHOD -1
j =0
for i in expense:
    if i == 2000:
        print("2000 spent at index ",j)
        break
    j += 1
else:
    print("Exact 2000 not spent in any month")

# METHOD -2
print("Did I spent 2000$ in any month? ", 2000 in expense)

# METHOD -3
for index, value in enumerate(expense):  # Best alternate way
    if value == 2000:
        print("2000 spent at index", index)
        break
else:
    print("Exact 2000 not spent in any month")
```

```python
# June month just finished and your expense is 1980 dollar. Add this item to our monthly expense list

expense.append(1980)  #add expense for june, will be added at last
print("Updated expense list month-wise (Jan-June):", expense)
```

```python
# You returned an item that you bought in a month of April and got a refund of 200$.
# Make a correction to your monthly expense list based on this

expense[3] = expense[3]-200
print("Updated expense after removing refund):", expense)
```

#### Exercise - 3
Create a list of all odd numbers between 1 and a max number. Max number is something you need to take from a user using input() function


```python
# Method - 1
max_num = int(input("Enter the max number to generate odd numbers list:"))

i= 1
odd_num=[]
while i<max_num:
    if i%2 != 0:
        odd_num.append(i)
    i+=1

print(odd_num)
```

```python
# Method - 2
max = int(input("Enter max number: "))
odd_numbers = []
for i in range(1, max):
    if i % 2 == 1:
        odd_numbers.append(i)
print("Odd numbers: ", odd_numbers)
```
