## Queue
A queue is a linear data structure that follows: FIFO principle (First in First Out).

Meaning -  The element inserted first is removed first.

Use Scenarios:
1) CPU Task Scheduling:
    - Process arrive - P1 -> P2 -> P3
    - Execution - P1 first -> P2 Second -> P3 third

2) Printer Queue
    - Documents - Doc1 -> Doc2 -> Doc 3
    - Printed - Doc1 -> Doc2 -> Doc 3

#### Implement queue class using collections.deque


```python
from collections import deque   # deque = Double Ended Queue.
                                # Optimized for fast insertion and deletion
                                # from both left and right ends (O(1)).

class Queue:

    def __init__(self):  # Constructor runs automatically when Queue object is created.
        self.buffer = deque()  # Create an empty deque to store queue elements.

    def enqueue(self, val):
        # Enqueue = Insert element into queue.
        # appendleft() inserts at the LEFT side.  We are treating LEFT as the REAR of the queue.
        self.buffer.appendleft(val)

    def dequeue(self):
        # Dequeue = Remove element from queue.
        # pop() removes from the RIGHT side. We are treating RIGHT as the FRONT of the queue.
        # FIFO (First In First Out) is maintained.
        return self.buffer.pop()

    def is_empty(self):
        # Returns True if queue contains no elements.  Returns False otherwise.
        return len(self.buffer) == 0

    def size(self):
        # Returns total number of elements in queue.
        return len(self.buffer)
```

```python
pq = Queue()

pq.enqueue({              # Insert element into queue
    'company': 'Wall Mart',
    'timestamp': '15 apr, 11.01 AM',
    'price': 131.10
})
pq.enqueue({
    'company': 'Wall Mart',
    'timestamp': '15 apr, 11.02 AM',
    'price': 132
})
pq.enqueue({
    'company': 'Wall Mart',
    'timestamp': '15 apr, 11.03 AM',
    'price': 135
})
```

```python
print(pq.buffer)    # Print the underlying deque
```

```python
print(list(pq.buffer))   # Print as a list
```

```python
pq.size()
```

```python
# FIFO
pq.dequeue()  #return and remove top element (rightmost element)
```

```python
pq.dequeue()
```

```python
pq.is_empty()
```
